```python
df_pe2019['TP_COR_RACA'] = df_pe2019['TP_COR_RACA'].replace({0: 'Não declarado',
                                                            1: 'Branca',
                                                            2: 'Preta',
                                                            3: 'Parda',
                                                            4: 'Amarela',
                                                            5: 'Indígena'})
```

```python
df_pe2019['TP_ESCOLA'] = df_pe2019['TP_ESCOLA'].replace({1: 'Não respondeu',
                                                        2: 'Público',
                                                        3: 'Privada',
                                                        4: 'Exterior'})
```

```python
df_pe2019['TP_LINGUA'] = df_pe2019['TP_LINGUA'].replace({0: 'Inglês',
                                                         1: 'Espanhol'})
```

### Separar entre vestibulandos e treineiros

> Filtrar apenas para os inscritos considerados vestibulandos e que não estão com o intuito de apenas treinar seus conhecimentos.

```python
vestibulandos = df_pe2019.loc[df_pe2019.IN_TREINEIRO == 'Não']
```

### Calcular a presença total dos inscritos

> Filtrar apenas para os inscritos, considerados vestibulandos, que participaram dos dois dias de vestibular.

```python
presenca_total = ['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']
```

```python
vestibulandos['Presença Total'] = vestibulandos[presenca_total].sum(axis=1)
```

```python
vestibulandos_presentes = vestibulandos.loc[vestibulandos['Presença Total'] == 4]
```

### Separar os vestibulandos classificados

> Apenas nos interessa os vestibulandos que fizeram todas as provas nos dois dias que compareceram ao vestibular. Ou seja, aqueles que não tiveram nenhum tipo de erro.

```python
vestibulandos_classificados_2019 = vestibulandos_presentes.loc[vestibulandos_presentes.TP_PRESENCA_MT == 1]
```
